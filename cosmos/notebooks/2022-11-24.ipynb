````python
from spacy.matcher import Matcher, DependencyMatcher
matcher = Matcher(nlp.vocab)

doc = nlp("shops within 1 mile of San Francisco?")

NOUN_TYPES = ["NOUN", "PROPN"]

matcher = DependencyMatcher(nlp.vocab) 
# {trajector} [within|in] {distance} of {landmark}
matcher.add("dist-1", [[
    # Root
    {
        "RIGHT_ID": "indicator",
        "RIGHT_ATTRS": {"LEMMA": {"IN": ["within", "in", "inside"]}, "POS": "ADP", "DEP": "prep"},        
    },
    {
        "RIGHT_ID": "dist",
        "LEFT_ID": "indicator",
        "REL_OP": ">",
        "RIGHT_ATTRS": {"ENT_TYPE": "DISTANCE", "DEP": "pobj"},
    },
    {
        "RIGHT_ID": "of",
        "LEFT_ID": "dist",
        "REL_OP": ">",
        "RIGHT_ATTRS": {"POS": "ADP", "DEP": "prep"},
    },   
    {
        "RIGHT_ID": "landmark",
        "LEFT_ID": "of",
        "REL_OP": ">",
        "RIGHT_ATTRS": {"POS": {"IN": NOUN_TYPES}, "DEP": "pobj"},
    }
]])

def find_trajector(tok):
    # Case: "shops in prep"
    # Try to find the trajector to the left of the spatial relationship
    trajector = None
    while True:
        tok = tok.head
        if tok.pos_ in NOUN_TYPES:
            return tok
        if tok.dep_ == "ROOT":
            if tok.pos_ == "AUX":
                for child in tok.children:
                    if child.dep_ == "nsubj":
                        return child                    
            if tok.pos_ == "VERB":
                for child in tok.children:
                    if child.dep_ in {"dobj", "nsubjpass"} and child.pos_ in NOUN_TYPES:
                        return child
            break
        if tok.head == tok:
            break
    return None

# doc = nlp("shops in 1 mile of San Francisco?")
# doc = nlp("shops that are in 1 mile of San Francisco?")
# doc = nlp("shops that are 1 mile of San Francisco?")
#doc = nlp("where are shops within 1 mile of San Francisco?")
#doc = nlp("what shops are within 1 mile of San Francisco?")
#doc = nlp("what shops are located within 1 mile of San Francisco?")
# doc = nlp("what shops are there that are located within 1 mile of San Francisco?")
# doc = nlp("give me shops within 1 mile of San Francisco?")
# doc = nlp("search for shops within 1 mile of San Francisco?")
# doc = nlp("What are the shops within 1 mile of San Francisco?")
for m_id, tokens in matcher(doc):
    print(doc.vocab.strings[m_id], [doc[i] for i in tokens], find_trajector(doc[tokens[0]]))

displacy.render(doc, style="dep")
````

````python
import spacy

KILOMETERS = ["km", "kilometers", "kilometer", "kilometres", "kilometre"]
METERS = ["m", "meters", "meter", "metres", "metre"]
MILES = ["mi", "miles", "mile"]
DIST_UNITS = KILOMETERS + METERS + MILES

dist_patterns = [
    [
     {"LIKE_NUM": True},
     {"LOWER": {"IN": DIST_UNITS}}
    ]
]

HOURS = ["hours", "hour", "hrs", "hr", "h"]
MINUTES = ["minutes", "minute", "mins", "min", "m"]
SECONDS = ["seconds", "second", "secs", "sec", "s"]

time_patterns = [
    *[[
        {"LIKE_NUM": True, "ENT_TYPE": "TIME"},
        {"LOWER": {"IN": unit}}
    ] for unit in [HOURS, MINUTES, SECONDS]],
    *[[
        {"LIKE_NUM": True, "ENT_TYPE": "TIME"},
        {"LOWER": {"IN": unit1}},
        {"POS": "CCONJ", "OP": "?"},
        {"LIKE_NUM": True, "ENT_TYPE": "TIME"},
        {"LOWER": {"IN": unit2}}
    ] for (unit1, unit2) in [(HOURS, MINUTES), (MINUTES, SECONDS), (HOURS, SECONDS)]],
    *[[
        {"LIKE_NUM": True, "ENT_TYPE": "TIME"},
        {"LOWER": {"IN": unit1}},
        {"POS": "CCONJ", "OP": "?"},
        {"LIKE_NUM": True, "ENT_TYPE": "TIME"},
        {"LOWER": {"IN": unit2}},
        {"POS": "CCONJ", "OP": "?"},
        {"LIKE_NUM": True, "ENT_TYPE": "TIME"},
        {"LOWER": {"IN": unit3}}
    ] for (unit1, unit2, unit3) in [(HOURS, MINUTES, SECONDS)]],
    [{"TEXT": {"REGEX": "^[0-9]{1,2}:[0-9]{2}(:[0-9]{2})?$"}, "ENT_TYPE": "TIME"}],
    [
        {"LIKE_NUM": True},
        {"LOWER": "h"},
        {"POS": "CCONJ", "OP": "?"},
        {"LIKE_NUM": True},
        {"LOWER": "m"}
    ],
    [
        {"LIKE_NUM": True},
        {"LOWER": "m"},
        {"POS": "CCONJ", "OP": "?"},
        {"LIKE_NUM": True},
        {"LOWER": "s"}
    ],
    [
        {"LIKE_NUM": True},
        {"LOWER": "h"},
        {"POS": "CCONJ", "OP": "?"},
        {"LIKE_NUM": True},
        {"LOWER": "s"}
    ],
]

nlp = spacy.load("en_core_web_lg")
# Add an entity ruler
entity_ruler = nlp.add_pipe("entity_ruler", after="ner", config={"overwrite_ents": True})
entity_ruler.add_patterns([{"label": "DURATION", "pattern": p} for p in time_patterns]) # type: ignore
entity_ruler.add_patterns([{"label": "DISTANCE", "pattern": p} for p in dist_patterns]) # type: ignore

# Merge entities
nlp.add_pipe("merge_entities")

````
